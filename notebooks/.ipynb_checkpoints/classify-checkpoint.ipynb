# **2. Import Data** 

```python
data_path = r'C:/Users/muhab/Sawit_Pro_2/data/classify/data.yaml'  
processed_folder = r'C:/Users/muhab/Sawit_Pro_2/data/classify/Processed/'
raw_images_folder = r'C:/Users/muhab/Sawit_Pro_2/data/classify/raw/'
os.makedirs(processed_folder, exist_ok=True)
```

# **3. Train Model** 

```python
model = YOLO("yolov8m.pt")
```

```python
model.train(
    data=data_path, 
    epochs=300,            
    imgsz=224,             
    batch=8,              
    conf=0.2,             
    optimizer='AdamW',     # AdamW is good, but consider SGD for better generalization
    lr0=0.001,             # Initial learning rate
    lrf=0.01,              # Final learning rate (cosine scheduler will gradually reduce)
    momentum=0.9,          # For SGD (ignored for AdamW)
    weight_decay=1e-4,     # Regularization (L2 penalty) to prevent overfitting
    dropout=0.2,           # Add dropout if your model supports it
    patience=20,           # Early stopping if no improvement after 20 epochs
    device='cuda',         
    save=True
)
```

```python
metrics = model.val()
print("Evaluation Metrics:", metrics)
```

# **4. Test in Valida Folder** 

```python
import matplotlib.pyplot as plt
import glob

# Locate saved validation images
#val_images = glob.glob('C:/Users/muhab/data/count/valid/*/*.jpg')  # Update path if necessary → C:\Users\muhab\Sawit_Pro_2\data\count\valid\images
val_images = glob.glob(r'C:/Users/muhab/Sawit_Pro_2/data/classify/valid/images/*.jpg')

if val_images:
    fig, axes = plt.subplots(nrows=len(val_images)//2 + 1, ncols=2, figsize=(50, 30))
    axes = axes.flatten()

    for idx, image_path in enumerate(val_images):
        eval_image = cv2.imread(image_path)
        eval_image = cv2.cvtColor(eval_image, cv2.COLOR_BGR2RGB)  # Convert to RGB for Matplotlib
        
        axes[idx].imshow(eval_image)
        axes[idx].axis('off')
        axes[idx].set_title(f"Eval Image {idx+1}")

    # Hide unused subplots if any
    for j in range(idx + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()
else:
    print("No evaluation images found.")
```

```python
for image_file in os.listdir(raw_images_folder):
    if not image_file.lower().endswith((".jpg", ".jpeg", ".png")):
        continue  

    image_path = os.path.join(raw_images_folder, image_file)
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not load {image_file}")
        continue

    # Run YOLO inference with tuned parameters
    results = model(image_path, conf=0.35, imgsz=320, iou=0.05, augment=True, max_det=10000)

    # Save processed image with bounding boxes
    processed_image_path = os.path.join(processed_folder, f"processed_{image_file}")

    # Draw bounding boxes with adjusted settings
    img_with_boxes = results[0].plot(line_width=2, font_size=5)  # Adjust box & font size
    cv2.imwrite(processed_image_path, img_with_boxes)

    # Convert to RGB for Matplotlib display
    img_with_boxes = cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB)

    # Display image with detections
    #plt.figure(figsize=(100,80))
    plt.imshow(img_with_boxes)
    plt.axis("off")
    plt.title(f"Processed: {image_file}")
    plt.show()

    # Count detected trees
    num_apples = len(results[0].boxes)
    print(f"{image_file}: Detected {num_apples} Apple")
```

# **5. Cropped Image** 

```python

os.makedirs(processed_folder, exist_ok=True)  # Ensure the cropped folder exists

# Initialize apple count for each color
apple_counts = {"red": 1, "yellow": 1, "green": 1}

# YOLO detection loop
for image_file in os.listdir(raw_images_folder):
    if not image_file.lower().endswith((".jpg", ".jpeg", ".png")):
        continue  

    image_path = os.path.join(raw_images_folder, image_file)
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not load {image_file}")
        continue

    # Run YOLO inference with tuned parameters
    results = model(image_path, conf=0.35, imgsz=320, iou=0.05, augment=True, max_det=10000)

    # Process detected apples
    for i, (box, cls) in enumerate(zip(results[0].boxes.xyxy, results[0].boxes.cls)):  
        x1, y1, x2, y2 = map(int, box.tolist())

        # Crop apple
        apple_crop = img[y1:y2, x1:x2]

        # Get apple class from YOLO
        apple_class = int(cls)  # YOLO class index
        class_names = ["green", "red", "yellow"]  # Make sure this matches your trained classes
        apple_color = class_names[apple_class]

        # Save cropped apple
        apple_filename = f"{apple_color}_{apple_counts[apple_color]}.jpg"
        apple_counts[apple_color] += 1
        apple_save_path = os.path.join(processed_folder, apple_filename)
        cv2.imwrite(apple_save_path, apple_crop)
        print(f"Saved: {apple_save_path}")

    # Save processed image with bounding boxes
    processed_image_path = os.path.join(processed_folder, f"processed_{image_file}")
    img_with_boxes = results[0].plot(line_width=2, font_size=5)
    cv2.imwrite(processed_image_path, img_with_boxes)

    # Convert for Matplotlib display
    img_with_boxes = cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB)
    plt.imshow(img_with_boxes)
    plt.axis("off")
    plt.title(f"Processed: {image_file}")
    plt.show()

    # Count detected apples
    num_apples = len(results[0].boxes)
    print(f"{image_file}: Detected {num_apples} Apple(s)")
```
